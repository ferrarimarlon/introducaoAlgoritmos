## 1. Proposições

Uma proposição é uma declaração que pode ser verdadeira ou falsa, mas não ambas. Exemplos de proposições incluem:

- "A Terra é redonda."
- "2 + 2 = 4."

## 2. Operadores Lógicos

### a. Negação (NOT)
A negação de uma proposição \( p \) é verdadeira se \( p \) é falsa e falsa se \( p \) é verdadeira.
- **Símbolo:** \( &not; p \)
- **Exemplo:** Se \( p \) é "A Terra é redonda", então \( &not; p \) é "A Terra não é redonda."

### b. Conjunção (AND)
O resultado \( p \) e \( q \) é verdadeiro somente se ambos \( p \) e \( q \) são verdadeiros.
- **Símbolo:** \( p &and; q \)
- **Exemplo:** "Hoje é segunda-feira" e "Está chovendo."

### c. Disjunção (OR)
O resultado \( p \) e \( q \) é verdadeiro se pelo menos um deles é verdadeiro.
- **Símbolo:** \( p &or; q \)
- **Exemplo:** "Hoje é segunda-feira" ou "Está chovendo."

## 3. Tabela Verdade

Uma tabela verdade é uma tabela que mostra todos os valores possíveis para uma expressão lógica. Por exemplo, para \( p \) e \( q \):

| \( p \)  | \( q \)  | \( &not; p \) | \( &not; q \) | \( p &and; q \) | \( p &or; q \) |

| V       | V       | F            | F            | V               | V              |
| V       | F       | F            | V            | F               | V              |

## 4. Leis Fundamentais da Lógica

### a. Lei da Identidade
Qualquer proposição é idêntica a si mesma.
- \( p &equals; p \)

### b. Lei da Não-Contradição
Nenhuma proposição pode ser verdadeira e falsa ao mesmo tempo.
- \( &not; (p &and; &not; p) \)

### c. Lei do Terceiro Excluído
Toda proposição ou é verdadeira ou é falsa.
- \( p &or; &not; p \)

# PARTE 02: Teoria de Conjuntos

### 1. Conjunto
Um conjunto é uma coleção de objetos distintos, chamados elementos, que são agrupados em uma única entidade. Por exemplo:
- O conjunto de todos os números pares menores que 10: {2, 4, 6, 8\}
- O conjunto das vogais do alfabeto: {a, e, i, o, u\}

```python
# definindo um conjunto
A = {2,4,6,8}
compras = {'TRIGO','CARNE','LEITE'}
```

### 2. Elemento
Um elemento é um dos objetos que fazem parte de um conjunto. Escrevemos \( x &in; A \) para indicar que o elemento \( x \) pertence ao conjunto \( A \). Por exemplo:
- \( 3 &in; \{1, 2, 3, 4, 5\} \) porque o número 3 está no conjunto.
- \( b &notin; \{a, c, d, e\} \) porque a letra b não está no conjunto.

```python
# verificando se um elemento pertence a um conjunto
3 in {1,2,3,4,5}
```

```python
'b' not in {'a','c','d','e'}
```

### 3. Subconjunto
Um conjunto \( B \) é um subconjunto de um conjunto \( A \) se todos os elementos de \( B \) também pertencem a \( A \). Escrevemos \( B &subset; A \) para indicar que \( B \) é um subconjunto de \( A \). Por exemplo:
- \( \{2, 4, 6\} &subset; \{1, 2, 3, 4, 5, 6\} \)
- \( \{a, b\} &subset; \{a, c, d\} \)

```python
{2, 4, 6}.issubset({1, 2, 3, 4, 5, 6})
```

```python
{'a', 'b'}.issubset({'a', 'c', 'd'})
```

### 4. Conjunto Vazio
O conjunto vazio, denotado por \( &emptyset; \) ou \( \{\} \), é um conjunto que não contém nenhum elemento. É importante notar que o conjunto vazio é um subconjunto de qualquer conjunto. Por exemplo:
- \( &emptyset; &subset; \{1, 2, 3\} \)

```python
# set() indica um conjunto vazio
set().issubset({1, 2, 3})
```

### 5. Conjunto Universo
O conjunto universal, denotado por \( U \), é o conjunto que contém todos os elementos relevantes para uma discussão particular. Por exemplo:
- Se estamos discutindo números inteiros, o conjunto universal poderia ser o conjunto dos inteiros.
- Se estamos discutindo cores, o conjunto universal poderia ser o conjunto de todas as cores possíveis.

## 6. Operações em Conjuntos

### 1. União &cup;
A união de dois conjuntos \( A \) e \( B \), denotada por \( A &cup; B \), é o conjunto que contém todos os elementos que estão em \( A \), em \( B \), ou em ambos.
- Exemplo: Se \( A = \{1, 2, 3\} \) e \( B = \{3, 4, 5\} \), então \( A &cup; B = \{1, 2, 3, 4, 5\} \).

```python
A = {1, 2, 3}
B = {3, 4, 5}
A.union(B)
```

### 2. Interseção &cap;
A interseção de dois conjuntos \( A \) e \( B \), denotada por \( A &cap; B \), é o conjunto que contém todos os elementos que estão em ambos \( A \) e \( B \).
- Exemplo: Se \( A = \{1, 2, 3\} \) e \( B = \{3, 4, 5\} \), então \( A &cap; B = \{3\} \).

```python
A = {1, 2, 3}
B = {3, 4, 5}
A.intersection(B)
```

### 3. Diferença -
A diferença entre dois conjuntos \( A \) e \( B \), denotada por \( A - B \), é o conjunto que contém todos os elementos de \( A \) que não estão em \( B \).
- Exemplo: Se \( A = \{1, 2, 3\} \) e \( B = \{3, 4, 5\} \), então \( A - B = \{1, 2\} \).

```python
A = {1, 2, 3}
B = {3, 4, 5}
A - B
```

## Intervalos

### 1. Intervalo Fechado
Um intervalo fechado inclui todos os números entre e incluindo seus limites. É denotado por [a, b], onde a é o limite inferior e b é o limite superior.
- Exemplo: O intervalo fechado de 1 a 5 é [1, 5], que inclui os números 1, 2, 3, 4 e 5.

```python
# representando [1,5]
intervalo_fechado = {1,2,3,4,5}
```

### 2. Intervalo Aberto
Um intervalo aberto inclui todos os números entre, mas não incluindo, seus limites. É denotado por (a, b) , onde a  é o limite inferior e b  é o limite superior.
- Exemplo: O intervalo aberto de 1 a 5 é (1, 5), que inclui os números entre 1 e 5, mas não inclui 1 e 5.

```python
# representando (1,5)
intervalo_aberto = {2,3,4}
```

### 3. Intervalo Semifechado
Um intervalo semifechado inclui um limite e exclui o outro.
Pode ser superiormente semifechado [a, b)  ou inferiormente semifechado (a, b]
- Exemplo: O intervalo superiormente semifechado de 1 a 5 é [1, 5), que inclui 1, 2, 3 e 4, mas não inclui 5.

```python
# representando [1,5)
intervalo_semifechado = {1,2,3,4}
# representando (1,5]
intervalo_semifechado = {2,3,4,5}
```

### Método range()

No Python, o método Range() é utilizado para criar intervalos. Sua característica é ser um intervalo semifechado [..), ou seja, não englobando o último elemento.

```python
# criando um intervalo [0,5) com Range()
intervalo = range(0,5)
# exibindo na tela em conjunto
set(intervalo)
```

O range() possui três parâmetros:
* início: o primeiro valor a ser adicionado no intervalo
* fim: o intervalo aberto do valor a ser adicionado, ou seja, o valor anterior a este
* incremento: qual a razão de incremento do intervalo (1 em 1, 2 em 2)

```python
# criando um conjunto dos 10 primeiros numeros pares
# lembrar que range(inicio, fim-1, incremento)
intervalo = range(0,20,2)
set(intervalo)
```

## Similaridade de Jaccard

A similaridade (ou índice) de Jaccard mede o quão próximos estão dois conjuntos. Sua escala inicia em 0, que significa que são totalmente diferentes (não possuem similaridade) e 1 que são idênticos.
Sua fórmula é dada pela razão entre os elementos em comum (interseção) e todos os elementos de ambos (união)

image.png

```python
ana = {'Batata','Frango','Alface','Whey'}
pedro = {'Whey','Frango','Alface'}
# calculando o quão próximas estão as dietas de ana e pedro
# len() pega o total de elementos para a divisão
jaccard = len(pedro.intersection(ana)) / len(pedro.union(ana))
jaccard
```

No Caso acima, a similaridade entre Ana e Pedro é de 75%, ou seja, uma alta similaridade.

No caso abaixo, imagine que Lucas tenha escrito o seguinte texto:
* "Os resultados do meu estudo permitem concluir que é possível uma diferença entre a data real do ocorrido histórico e seu registro de fato."
Já Joel escreveu o seguinte:
* "Os resultados do estudo permitem concluir uma diferença entre a data do ocorrido histórico e seu registro."

Um professor de história, desconfiando de um caso de plágio, deseja saber se existe similaridade entre os dois textos. Assim, ele utiliza a similaridade de Jaccard, transformando cada palavra em um elemento de conjunto.

```python
Lucas = {"Os","resultados","do","meu","estudo","permitem","concluir","que","é","possível","uma","diferença","entre","a","data","real","do","ocorrido","histórico","e","seu","registro","de","fato."}
Joel = {"Os","resultados","do","estudo","permitem","concluir","uma","diferença","entre","a","data","do","ocorrido","histórico","e","seu","registro"}
jaccard = len(Lucas.intersection(Joel)) / len(Lucas.union(Joel))
jaccard
```

Com 70% de similaridade entre os dois textos, é possível comprovar um caso de plágio entre os dois alunos.

### Exemplo 4

Unificando os conceitos de conjuntos, lógica e estruturas relacionais, conseguimos construir algoritmos lógicos de forma eficiente. Vejamos este exemplo:

Uma plataforma de comércio online deseja disparar um email marketing a usuários com perfis da campanha "Mês Fit". Na campanha, usuários que possuem um perfil de consumo de alimentos saudáveis receberá um cupom de 30% de desconto na compra. Os alimentos saudáveis da loja são:  'maçã', 'banana', 'espinafre', 'aveia', 'cenoura', 'iogurte', 'salmão', 'brócolis' e 'quinoa'

A área de marketing verificou que Lucas pode ser um potencial cliente a esta campanha, e por isso resolveu listar os últimos produtos comprados por ele: 'espinafre', 'aveia', 'cenoura', 'iogurte', 'salmão' e 'brócolis'

Diante destas informações, crie uma estrutura lógica em algoritmo que calcule se Lucas é um usuário similar para esta campanha. Para ser considerado elegível, a similaridade precisa ser de pelo menos 60%.

```python
publico_alvo = { 'maçã', 'banana', 'espinafre', 'aveia', 'cenoura', 'iogurte', 'salmão', 'brócolis', 'quinoa'}
lucas = {'espinafre', 'aveia', 'cenoura', 'iogurte', 'salmão', 'brócolis'}
# calcula o indice de jaccard para similaridade
similaridade = len(publico_alvo.intersection(lucas)) / len(publico_alvo.union(lucas))
similaridade
```

Com similaridade de 66%, Lucas é um candidato à campanha. Uniremos o requisito de pelo menos 60% de similaridade com uma mensagem de desconto a ele:

```python
'Você ganhou desconto de 30% na semana Fit!' * (similaridade >=0.6)
```

Entretanto, as diretrizes de privacidade dizem que um usuário só poderá receber emails se ele está cadastrado na base de informativos. Assim, precisamos inserir também esta condição:

```python
quer_receber_email = True
'Você ganhou desconto de 30% na semana Fit!' * (similaridade >=0.6 and quer_receber_email)
```

Neste caso, o valor foi exibido, uma vez que a conjunção foi Verdadeira. Caso Lucas optasse por não receber emails, seu valor seria Falso e a mensagem não seria exibida.

# PARTE 04: Estruturas de Iterações

As estruturas de iteração, ou loops, são um conceito fundamental em programação, permitindo a execução repetida de um bloco de código enquanto uma condição específica é verdadeira. Essas estruturas são essenciais para automatizar tarefas repetitivas, processar coleções de dados e controlar o fluxo do programa de maneira eficiente.
Em Python, as principais estruturas de iteração são os loops for e while.
Utilizar repetições amplia o escopo dos algoritmos, tornado-os capazes de processar milhões de registros.

### FOR

O loop for é utilizado para iterar sobre uma sequência (como listas, tuplas, dicionários, conjuntos ou strings) ou qualquer objeto que seja iterável.
Ele percorre cada item da sequência, executando um bloco de código para cada elemento.

```python
# percorrendo e exibindo cada elemento da lista
frutas = ["jaca", "mamão", "laranja"]
for fruta in frutas:
    print(fruta)
```

O FOR pode ser utilizado com o conceito de intervalos para criar repetições baseadas na quantidade de itens de uma lista.

```python
# o resultado abaixo é o mesmo do código anterior
# porém percorre cada índice da lista em vez de cada elemento
frutas = ["jaca", "mamão", "laranja"]
for i in range(len(frutas)):
    print(frutas[i])
```

#### List Comprehension

Uma forma concisa e eficiente de criar listas em Python é usando list comprehensions, que utilizam a estrutura do for dentro de uma expressão compacta.
Permitem a criação de listas de maneira mais legível e com menos código.

```python
# list comprehension com os 10 primeiros números pares
pares = [numero for numero in range(0,20,2)]
pares
```

```python
# mesmo código anterior, sem list comprehension
pares = []
for numero in range(0,20,2):
  pares.append(numero)
pares
```

# PARTE 05: Distâncias e Similaridades

### Métricas de Distância e Similaridade

Métricas de distância e similaridade são ferramentas essenciais em diversos algoritmos, como aprendizado de máquina, mineração de dados, análise de redes sociais, recuperação de informações e bioinformática. Essas métricas são utilizadas para quantificar a proximidade ou similaridade entre objetos, que podem ser pontos em um espaço, vetores de características, sequências de texto, entre outros.


#### Distância Euclidiana

Baseada no teorema de pitágoras, é a hipotenusa entre dois pontos em um espaço n-dimensional.

image.png

Nas métricas de distância, como a euclidiana, quanto mais próximo de zero o valor, mais similares os dois pontos.

##### Exemplo 1

João e Lucas são dois clientes da empresa "SOS Beleza". No mês passado, João comprou 12 itens cosméticos e gastou um total de R\$ 5.200,00. Já Lucas comprou 11 itens e gastou um total de R\$4.500,00. Calcule a distância entre os dois consumidores:

```python
joao = [12,5200]
lucas = [11,4500]

# elevar um número a 0.5 é tirar sua raiz quadrada
distancia = ((joao[0]-lucas[0])**2 + (joao[1]-lucas[1])**2)**0.5
distancia
```

Ao utilizar listas, é possível tornar a operação mais enxuta utilizando o FOR:

```python
joao = [12,5200]
lucas = [11,4500]

soma_catetos = 0
for i in range(len(joao)):
  soma_catetos = soma_catetos + (joao[i]-lucas[i])**2
#raiz quadrada
distancia = soma_catetos**0.5
distancia
```

Com um valor de 700 de distância, podemos considerar ambos similares, porém a distância euclidiana é altamente sensível à escala dos dados. Como neste caso em que estamos comparando total de itens e valores monetários, a escala é muito maior para os valores (milhares de reais), tornando a interpretação dos resultados mais difícil. Para evitar que os dados estejam em escalas diferentes no plano cartesiano, uma técnica importante é aplicar uma normalização de escala antes de calcular a distância euclidiana.

##### Normalização de escalas

A normalização de escalas permite que todos os valores de uma lista, conjunto ou vetor sejam reposicionados em um mesmo intervalo, normalmente entre 0 e 1, garantindo a consistência nas fórmulas de distância, por todos os dados estarem padronizados. Uma destas fórmulas é o minmax:

image.png

```python
lista1 = [1000,1,50]
lista1_comEscala = []

X_max = max(lista1)
X_min = min(lista1)

for X in lista1:
  Xnovo = (X-X_min)/(X_max-X_min)
  lista1_comEscala.append(Xnovo)
lista1_comEscala
```

Assim, uma lista heterogênea [1000,1,50] torna-se [1.0, 0.0, 0.04904904904904905] em uma nova escala de 0 a 1.

##### Exemplo 2

Uma aplicação importante da normalização é quando estamos comparando escalas diferentes, como por exemplo uma lista está em celsius e outra em fahrenheit, ou uma lista está em kilômetros e a outra em metros.

```python
# calculando a distância de dois fornos industriais
forno_celsius = [176.67, 190.56, 204.44, 218.33, 232.22, 246.11, 260.0, 273.89, 287.78, 301.67]
forno_fahrenheit = [350, 375, 400, 425, 450, 475, 500, 525, 550, 575]
```

As duas temperaturas são equivalentes, porém em escalas diferentes. Se ignorarmos este fato, a distância euclidiana não conseguirá compreender a proximidade destas distâncias, conforme exemplo:

```python
# calculando a distância SEM normalizar os dados
soma_catetos = 0
for i in range(len(forno_celsius)):
  soma_catetos = soma_catetos + (forno_celsius[i]-forno_fahrenheit[i])**2
#raiz quadrada
distancia = soma_catetos**0.5
distancia
```

Normalizando as duas listas de temperaturas:

```python
# normalizando celsius
forno_celsius_norm = []
X_max = max(forno_celsius)
X_min = min(forno_celsius)

for X in forno_celsius:
  Xnovo = (X-X_min)/(X_max-X_min)
  forno_celsius_norm.append(Xnovo)
forno_celsius_norm
```

```python
# normalizando celsius
forno_fahrenheit_norm = []
X_max = max(forno_fahrenheit)
X_min = min(forno_fahrenheit)

for X in forno_fahrenheit:
  Xnovo = (X-X_min)/(X_max-X_min)
  forno_fahrenheit_norm.append(Xnovo)
forno_fahrenheit_norm
```

Uma vez normalizados, percebemos que ambas as listas são idênticas em valores (ignorando as casas infinitesimais). Assim, calculando a distância euclidiana, provamos que elas são completamente similares:

```python
# calculando a distância COM os dados normalizados
soma_catetos = 0
for i in range(len(forno_celsius_norm)):
  soma_catetos = soma_catetos + (forno_celsius_norm[i]-forno_fahrenheit_norm[i])**2
#raiz quadrada
distancia = soma_catetos**0.5
distancia
```

### Similaridade do Cosseno

A similaridade do cosseno é uma métrica que mede a similaridade entre dois vetores não nulos. Essa métrica é amplamente utilizada em aprendizado de máquina, processamento de linguagem natural e recuperação de informações para comparar documentos ou perfis de usuários.

image.png

A fórmula para calcular a similaridade do cosseno entre dois vetores A e B é:

image.png

Na parte do nominador, o conceito utilizado é o produto de pontos de dois vetores. Já na parte denominadora, temos o teorema de pitágoras.

##### Produto de pontos

O produto de pontos, também conhecido como produto escalar, é uma operação matemática que toma dois vetores de igual dimensão e retorna um único número. Este número é o resultado da multiplicação correspondente dos componentes dos vetores, somados juntos.

image.png

```python
# implementação do produto de pontos
A = [1,2,3]
B = [4,5,6]
resultado = 0
for i in range(len(A)):
  resultado = resultado + A[i]*B[i]
resultado
```

#### Exemplo 1: Similaridade do Cosseno

No exemplo acima, calculamos o total a pagar de Jonas. Mas e se quiséssemos saber se Jonas e Pedro são clientes similares no mercado? Assim, podemos aplicar a similaridade do cosseno:

```python
itens_jonas = [3,1,12]
itens_pedro = [2,0,10]
#calculando o nominador da fórmula, produto dos pontos entre jonas e pedro
produto_pontos = 0
for i in range(len(itens_jonas)):
  produto_pontos = produto_pontos + itens_jonas[i]*itens_pedro[i]
```

Nesta primeira parte, fizemos o nominador da fórmula do cosseno:

image.png

Agora iremos para o denominador. Como são dois vetores, precisaremos calcular A primeiro (Jonas) e depois B (Pedro). A fórmula é a mesma, mundando apenas a lista.

```python
#calculando o denominador de jonas
hipotenusa_jonas = 0
for i in range(len(itens_jonas)):
  hipotenusa_jonas = hipotenusa_jonas + itens_jonas[i]**2
# tira a raiz
hipotenusa_jonas = hipotenusa_jonas**0.5

#calculando o denominador de pedro
hipotenusa_pedro = 0
for i in range(len(itens_pedro)):
  hipotenusa_pedro = hipotenusa_pedro + itens_pedro[i]**2
# tira a raiz
hipotenusa_pedro = hipotenusa_pedro**0.5
```

Para obtermos o resultado:

```python
produto_pontos / (hipotenusa_jonas*hipotenusa_pedro)
```

A escala de similaridade é oposta à da distância. Ou seja, enquanto para a distância, quanto mais próximo de 0 mais iguais, para similaridades quanto mais próximo de 1 mais iguais. Como Pedro e Jonas possuem similaridade de 0.995, indica uma altíssima similaridade entre eles.
